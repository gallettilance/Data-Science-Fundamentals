Outliers are points that lie far from the rest of the data. They are not necessarily invalid points however. Imagine sampling from a Normal Distribution with mean 10 and variance 1. You would expect most points you sample to be in the range [7, 13] but it's entirely possible to see 20 which, on average, should be very far from the rest of the points in the sample (unless we're VERY (un)lucky). These outliers can inhibit our ability to learn general patterns in the data since they are not representative of likely outcomes. They can still be useful in of themselves and can be analyzed in great depth depending on the problem at hand.

b) Are there any points in the dataset that could be outliers? If so, please remove them from the dataset.

Yes, there is one point that could be an outlier - the point in riht upper corner [10,10]

```python
a = np.array(data["Attribute A"])
b = np.array(data["Attribute B"])
classes = np.array(data["Class"])

def no_outliers(attributes, classes):
    mean = np.mean(attributes)
    std_dev = np.std(attributes)
    fil_indices = [i for i, value in enumerate(attributes) if (mean - 3 * std_dev < value < mean + 3 * std_dev)]
    fil_attributes = attributes[fil_indices]
    fil_classes = classes[fil_indices]
    return fil_attributes, fil_classes

a2, class2 = no_outliers(a, classes)
b2, class2 = no_outliers(b, classes)

data["Attribute A"] = a2.tolist()
data["Attribute B"] = b2.tolist()
data["Class"] = class2.tolist()

# print(data)

colors = np.array([x for x in 'bgrcmyk'])
plt.scatter(data["Attribute A"], data["Attribute B"], color=colors[data["Class"]].tolist())
plt.show() 
    
```

e) Write a function to compute the Euclidean distance from it to all points in the dataset and pick the 3 closest points to it. In a scatter plot, draw a circle centered around the point with radius the distance of the farthest of the three points.

```python
#function to find closest points
def n_closest_to(example, data, n):
    distances = np.sqrt(np.sum((data - example) ** 2, axis=1))
    iclosest = np.argsort(distances)[:n]
    closest_dist = distances[iclosest]
    radius = closest_dist[-1]
    return iclosest, radius

colors_filtered = np.array(['green' if c == 1 else 'blue' for c in data["Class"]])
data_points = np.array(list(zip(data["Attribute A"], data["Attribute B"])))
location = np.array([0.5, 1]) 
iclosest, radius = n_closest_to(location, data_points, 3)

_, axes = plt.subplots()
axes.scatter(data["Attribute A"], data["Attribute B"], color=colors_filtered)
axes.scatter(location[0], location[1], color='yellow')

for i in iclosest:
    axes.scatter(data_points[i][0], data_points[i][1], color='red')
cir = plt.Circle(location, radius, fill=True, color='grey', alpha=0.8)
axes.add_patch(cir)
axes.set_aspect('equal')
plt.show()
```

f) Write a function that takes the three points returned by your function in e) and returns the class that the majority of points have (break ties with a deterministic default class of your choosing). Print the class assigned to this new point by your function.

```python
def majority(points, data_classes):
    point_classes = [data_classes[i] for i in points]
    counts = np.bincount(point_classes)
    return np.argmax(counts)

assigned = majority(iclosest, data["Class"])
print("Class assigned to the new point is", assigned)
```

g) Re-using the functions from e) and f), you should be able to assign a class to any new point. In this exercise we will implement Leave-one-out cross validiation in order to evaluate the performance of our model.

For each point in the dataset:

- consider that point as your test set and the rest of the data as your training set
- classify that point using the training set
- keep track of whether you were correct with the use of a counter

Once you've iterated through the entire dataset, divide the counter by the number of points in the dataset to report an overall testing accuracy.

```python
count = 0
for i in range(len(data_points)):
    test_example = data_points[i]

    training_indices = list(range(len(data_points)))
    training_indices.remove(i)
    training_set = data_points[training_indices]

    iclosest, _ = n_closest_to(test_example, training_set, 3)
    prediction = majority(iclosest, classes[training_indices])
    
    if prediction == classes[i]:
        count += 1

overall_accuracy = count / len(data_points)
print("overall accuracy =", overall_accuracy)
```

## Challenge Problem

For this question we will re-use the "mnist_784" dataset.

a) Begin by creating a training and testing datasest from our dataset, with a 80-20 ratio, and random_state=1. You can use the `train_test_split` function from sklearn. By holding out a portion of the dataset we can evaluate how our model generalizes to unseen data (i.e. data it did not learn from).

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_openml

X, y = fetch_openml(name="mnist_784", version=1, return_X_y=True, as_frame=False)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

b) For K ranging from 1 to 20:

1. train a KNN on the training data
2. record the training and testing accuracy

Plot a graph of the training and testing set accuracy as a function of the number of neighbors K (on the same plot). Which value of K is optimal? Briefly explain.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt


train_acc = []
test_acc = []

for i in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)

    y_train_pred = knn.predict(X_train)
    y_test_pred = knn.predict(X_test)

    train_acc.append(accuracy_score(y_train, y_train_pred))
    test_acc.append(accuracy_score(y_test, y_test_pred))
    
k_values = range(1, 21)   
plt.figure(figsize=(12, 6))
plt.plot(k_values, train_acc, label='Training Accuracy')
plt.plot(k_values, test_acc, label='Testing Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.title('KNN Accuracy vs Number of Neighbors')
plt.legend()
plt.grid(True)
plt.show()

optimal = k_values[test_acc.index(max(test_acc))]
print(f"Optimal K: {optimal} with testing accuracy: {max(test_acc)}")
    
```

c) Using the best model from b), pick an image at random and plot it next to its K nearest neighbors

```python
import random

def n_closest_to(example, data, n):
    distances = np.sqrt(np.sum((data - example) ** 2, axis=1))
    closest_indices = np.argsort(distances)[:n]
    return closest_indices

best_knn = KNeighborsClassifier(n_neighbors=optimal)
best_knn.fit(X_train, y_train)

random_image = X_test[random.randint(0, len(X_test) - 1)]
neighbors_indices = n_closest_to(random_image.reshape(1, -1), X_train, optimal)
neighbors_images = X_train[neighbors_indices]

fig, axes = plt.subplots(1, optimal + 1, figsize=(10, 3))
axes[0].imshow(random_image.reshape(28, 28), cmap='summer')
axes[0].set_title("Random Image")
axes[0].axis('off')
for i, img in enumerate(neighbors_images):
    axes[i+1].imshow(img.reshape(28, 28), cmap='summer')
    axes[i+1].set_title(f"Neighbor {i+1}")
    axes[i+1].axis('off')

plt.show()
```

d) Using a dimensionality reduction technique discussed in class, reduce the dimensionality of the dataset before applying a KNN model. Repeat b) and discuss similarities and differences to the previous model. Briefly discuss your choice of dimension and why you think the performance / accuracy of the model has changed.

```python
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA


pca = PCA(n_components=0.95)
test_acc2 =[]
train_acc2 =[]

for i in range(1,21):
    knn_model = KNeighborsClassifier(n_neighbors=i)
    
    model = make_pipeline(pca, knn_model)
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    train_acc2.append(accuracy_score(y_train, y_train_pred))
    
    y_test_pred = model.predict(X_test)
    test_acc2.append(accuracy_score(y_test, y_test_pred))

plt.figure(figsize=(10, 6))
plt.plot(range(1,21), train_acc2, label='Training Accuracy')
plt.plot(range(1,21), test_acc2, label='Testing Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.title('KNN with PCA Accuracy vs Number of Neighbors')
plt.legend()
plt.show()

k_values = range(1,21)
optimal_k_pca = k_values[test_acc2.index(max(test_acc2))]
optimal_test_accuracy_pca = max(test_acc2)

print(f"Optimal K after PCA: {optimal_k_pca} with testing accuracy: {optimal_test_accuracy_pca}")
```

 Using PCA makes the model a little better to generalize as there is smoother decline in testing accuracy with increasing K.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

train_data = pd.read_csv('./train.csv')
test_data = pd.read_csv('./test.csv')

#combining the data so it is easier to add new features
combined= pd.concat([train_data, test_data], axis=0)
#feature Family Size, we expect people with families to have higher chance of survival than person traveling alone
combined['FamilySize'] = combined['SibSp'] + combined['Parch'] + 1 
#feature Age*Class, we expect to see younger people from higher classes to have better chance of survival 
combined['Age*Class'] = combined['Age'] * combined['Pclass'] 

#we drop unnecessary variables
combined.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

#set categorical and numerical columns
categorical_cols = ['Sex', 'Embarked']
numerical_cols = ['Pclass', 'Age', 'Fare', 'FamilySize', 'Age*Class']

#we make sure we have no missing values, 
#if we miss a value, we calculate the mean of the column and put it as a value
imputer = SimpleImputer(strategy='mean')
combined[numerical_cols] = imputer.fit_transform(combined[numerical_cols])

#scalling
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(), categorical_cols)])

#now we split back to train and test data
X_train = combined[:len(train_data)]
X_test = combined[len(train_data):]

#KNN pipeline definition
pipeline = Pipeline(steps=[('preprocessor', preprocessor),('knn', KNeighborsClassifier())])

#define parameter grid for GridSearchCV
param_grid = {'knn__n_neighbors': np.arange(1, 11)}

#perform GridSearchCV to find the best number for k
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, train_data['Survived'])

print("Best parameters:", grid_search.best_params_)
print("Best CV score:", grid_search.best_score_)

#now making prediction on the testing dataset
predictions = grid_search.predict(X_test)

#submission file preparation
submission = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': predictions})
submission.to_csv('titanic_submission_with_features.csv', index=False)


```

I decided to add 2 new features, one of them is **Family size** -  I combined 2 variables that mention whether the person is travelling alone (0) or with parents, and a second variable that says how many siblings the person and added +1 for the person itself. The other new feature is **Age times Class** - this is basically multiplication between column *Age* and *Class* which stands for person's age and their class (if higher class, then lower number). I expect to see that younger people from higher classes are more likely to survive, therefore the lower number of the new feature variable, the higher possibility of survival.

**Data Preparation:**

First, I prepare the data by handling missing values, encoding categorical variables, and scaling numerical features. This ensures that the data is in a format that the KNN algorithm can understand.

**Choosing the Number of Neighbors:**

KNN works by finding the "k" nearest data points to a given point (in this case, a passenger's characteristics) and using their labels (in this case, whether they survived or not) to make a prediction.
I use a technique called cross-validation to choose the best value of "k." I try different values of "k" and evaluate how well the model performs using each value through cross-validation. The value of "k" that gives the best performance is selected.

**Training the Model:**

Once I determine the best value of "k", I train the KNN model using the training data. During training, the model "learns" from the characteristics of passengers who survived and those who didn't.

**Making Predictions:**

After training, the model is ready to make predictions on new, unseen data (in this case, the test data). For each passenger in the test data, the model looks at the characteristics of the "k" nearest passengers in the training data and predicts whether the passenger survived or not based on the majority label among those "k" neighbors.
