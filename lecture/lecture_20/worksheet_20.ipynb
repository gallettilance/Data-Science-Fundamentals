## Gradient Descent

Recall in Linear Regression we are trying to find the line $$y = X \beta$$ that minimizes the sum of square distances between the predicted `y` and the `y` we observed in our dataset:

$$\mathcal{L}(\mathbf{\beta}) = \Vert \mathbf{y} - X\mathbf{\beta} \Vert^2$$

We were able to find a global minimum to this loss function but we will try to apply gradient descent to find that same solution.

a) Implement the `loss` function to complete the code and plot the loss as a function of beta.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

beta = np.array([ 1 , .5 ])
xlin = -10.0 + 20.0 * np.random.random(100)
X = np.column_stack([np.ones((len(xlin), 1)), xlin])
y = beta[0]+(beta[1]*xlin)+np.random.randn(100)

fig, ax = plt.subplots()
ax.plot(xlin, y,'ro',markersize=4)
ax.set_xlim(-10, 10)
ax.set_ylim(min(y), max(y))
plt.show()
```

```python
b0 = np.arange(-5, 4, 0.1)
b1 = np.arange(-5, 4, 0.1)
b0, b1 = np.meshgrid(b0, b1)

def loss(X, y, beta):
    predictions = X.dot(beta)
    errors = y - predictions
    return np.mean(errors ** 2)

def get_cost(B0, B1):
    res = []
    for b0, b1 in zip(B0, B1):
        line = []
        for i in range(len(b0)):
            beta = np.array([b0[i], b1[i]])
            line.append(loss(X, y, beta))
        res.append(line)
    return np.array(res)

cost = get_cost(b0, b1)
 
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
ax.set_xlim(-6, 4)
ax.set_xlabel(r'$\beta_0$')
ax.set_ylabel(r'$\beta_1$')
ax.set_ylim(-5, 4)
ax.set_zlim(0, 30000)

# Creating plot
ax.plot_surface(b0, b1, cost, alpha=.7)
 
# show plot
plt.show()
```

Since the loss is

$$\mathcal{L}(\mathbf{\beta}) = \Vert \mathbf{y} - X\mathbf{\beta} \Vert^2 = \beta^T X^T X \beta - 2\mathbf{\beta}^TX^T\mathbf{y}  + \mathbf{y}^T\mathbf{y}$$

The gradient is

$$\nabla_\beta \mathcal{L}(\mathbf{\beta}) = 2X^T X \beta - 2X^T\mathbf{y}$$

b) Implement the gradient function below and complete the gradient descent algorithm

```python
import numpy as np
from PIL import Image as im
import matplotlib.pyplot as plt

TEMPFILE = "temp.png"

def loss(X, y, beta):
    predictions = X.dot(beta)
    errors = y - predictions
    return np.mean(errors ** 2)

def snap(betas, losses):
    # Plot settings
    fig = plt.figure(figsize =(14, 9))
    ax = plt.axes(projection ='3d')
    ax.view_init(20, -20)
    ax.set_xlim(-5, 4)
    ax.set_xlabel(r'$\beta_0$')
    ax.set_ylabel(r'$\beta_1$')
    ax.set_ylim(-5, 4)
    ax.set_zlim(0, 30000)

    # Plot surface and path
    ax.plot_surface(b0, b1, cost, color='b', alpha=.7)
    ax.plot(np.array(betas)[:,0], np.array(betas)[:,1], losses, 'o-', c='r', markersize=10, zorder=10)
    fig.savefig(TEMPFILE)
    plt.close()
    return im.fromarray(np.asarray(im.open(TEMPFILE)))

def gradient(X, y, beta):
    return 2 * X.T.dot(X.dot(beta) - y)

def gradient_descent(X, y, beta_hat, learning_rate, epochs, images):
    losses = [loss(X, y, beta_hat)]
    betas = [beta_hat]

    for _ in range(epochs):
        images.append(snap(betas, losses))
        grad = gradient(X, y, beta_hat)
        beta_hat -= learning_rate * grad  # beta_hat and grad are both float64

        losses.append(loss(X, y, beta_hat))
        betas.append(beta_hat)
        
    return np.array(betas), np.array(losses)

# Initialize parameters with float64 type
beta_start = np.array([-5, -2], dtype=np.float64)
learning_rate = 0.0002  # try 0.0005
images = []

# Run gradient descent
betas, losses = gradient_descent(X, y, beta_start, learning_rate, 10, images)


# Create GIF
images[0].save(
    'gd.gif',
    optimize=False,
    save_all=True,
    append_images=images[1:],
    loop=0,
    duration=500
)
```

In logistic regression, the `loss` is the negative log-likelihood

$$ \mathcal{l}(\mathbf{\beta}) = - \frac{1}{N} \sum_{i=1}^{N} y_i \log(\sigma(x_i \beta)) + (1 - y_i)\log(1 - \sigma(x_i \beta))$$

the gradient of which is:

$$\nabla_\beta \mathcal{l}(\mathbf{\beta}) = \frac{1}{N} \sum_{i=1}^{N} x_i (y_i - \sigma(x_i \beta)) $$

```python
import numpy as np
from PIL import Image as im
import matplotlib.pyplot as plt

TEMPFILE = "temp.png"

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Logistic regression loss function
def loss(X, y, beta):
    z = X.dot(beta)
    sigmoid_values = sigmoid(z)
    return -np.mean(y * np.log(sigmoid_values) + (1 - y) * np.log(1 - sigmoid_values))

# Logistic regression gradient function
def gradient(X, y, beta):
    z = X.dot(beta)
    sigmoid_values = sigmoid(z)
    return X.T.dot(sigmoid_values - y) / len(y)

# Function to capture the state of the model
def snap(betas, losses):
    fig = plt.figure(figsize =(14, 9))
    ax = plt.axes(projection ='3d')
    ax.view_init(10, 10)
    ax.set_xlabel(r'$\beta_0$')
    ax.set_ylabel(r'$\beta_1$')
    ax.set_ylim(-20, 20)
    ax.set_zlim(0, 10)

    ax.plot_surface(b0, b1, cost, color='b', alpha=.7)
    ax.plot(np.array(betas)[:,0], np.array(betas)[:,1], losses, 'o-', c='r', markersize=10, zorder=10)
    fig.savefig(TEMPFILE)
    plt.close()
    return im.fromarray(np.asarray(im.open(TEMPFILE)))

# Gradient descent function
# Gradient descent function
def gradient_descent(X, y, beta_hat, learning_rate, epochs, images):
    losses = [loss(X, y, beta_hat)]
    betas = [beta_hat]

    for _ in range(epochs):
        images.append(snap(betas, losses))
        beta_hat -= learning_rate * gradient(X, y, beta_hat)  # beta_hat and gradient are both float64

        losses.append(loss(X, y, beta_hat))
        betas.append(beta_hat)
        
    return np.array(betas), np.array(losses)

# Initialize parameters with float64 type
beta_start = np.array([-5, -2], dtype=np.float64)
learning_rate = 0.1
images = []
betas, losses = gradient_descent(X, y, beta_start, learning_rate, 10, images)
```

g) Modify the above code to evaluate the gradient on a random batch of the data. Overlay the true loss curve and the approximation of the loss in your animation.

```python
def stochastic_gradient(X, y, beta, batch_size):
    indices = np.random.choice(len(y), batch_size, replace=False)
    X_batch = X[indices]
    y_batch = y[indices]
    z = X_batch.dot(beta)
    sigmoid_values = sigmoid(z)
    return X_batch.T.dot(sigmoid_values - y_batch) / len(y_batch)

def snap_model(beta, true_loss, approx_loss):
    xplot = np.linspace(np.min(X[:, 1]), np.max(X[:, 1]), 50)
    yplot = -(beta[0] + beta[1] * xplot) / beta[2]
    
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))

    # Plot the model
    ax[0].plot(xplot, yplot, 'b-', lw=2)
    ax[0].scatter(X[:, 1], X[:, 2], c=y, cmap='viridis')
    ax[0].set_xlim(np.min(X[:, 1]), np.max(X[:, 1]))
    ax[0].set_ylim(np.min(X[:, 2]), np.max(X[:, 2]))
    ax[0].set_title("Logistic Regression Fit")

    # Plot the losses
    ax[1].plot(true_loss, label='True Loss')
    ax[1].plot(approx_loss, label='Approximate Loss')
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("Loss")
    ax[1].legend()
    ax[1].set_title("Loss over Epochs")

    fig.savefig(TEMPFILE)
    plt.close()
    return im.fromarray(np.asarray(im.open(TEMPFILE)))

def gradient_descent(X, y, beta_hat, learning_rate, epochs, images, batch_size):
    true_losses = []
    approx_losses = []
    betas = [beta_hat]

    for _ in range(epochs):
        true_loss = loss(X, y, beta_hat)
        true_losses.append(true_loss)
        
        beta_hat -= learning_rate * stochastic_gradient(X, y, beta_hat, batch_size)
        approx_loss = loss(X, y, beta_hat)
        approx_losses.append(approx_loss)

        betas.append(beta_hat)
        images.append(snap_model(beta_hat, true_losses, approx_losses))
        
    return np.array(betas), np.array(true_losses), np.array(approx_losses)

# Add intercept term to X
X = np.column_stack([np.ones(X.shape[0]), X])

# Then proceed with the gradient descent as before
beta_start = np.array([1, -1, 1], dtype=np.float64)  # This is now correctly aligned with X
learning_rate = 0.1
batch_size = 20  # Adjust based on your dataset size
images = []

# Run gradient descent
betas, true_losses, approx_losses = gradient_descent(X, y, beta_start, learning_rate, 100, images, batch_size)

# Save the animation
images[0].save(
    'logistic_regression_sgd.gif',
    optimize=False,
    save_all=True,
    append_images=images[1:],
    loop=0,
    duration=200
)
```

h) Below is a sandox where you can get intuition about how to tune gradient descent parameters:

```python
import numpy as np
from PIL import Image as im
import matplotlib.pyplot as plt

TEMPFILE = "temp.png"

def snap(x, y, pts, losses, grad):
    fig = plt.figure(figsize =(14, 9))
    ax = plt.axes(projection ='3d')
    ax.view_init(20, -20)
    ax.plot_surface(x, y, loss(np.array([x, y])), color='r', alpha=.4)
    ax.plot(np.array(pts)[:,0], np.array(pts)[:,1], losses, 'o-', c='b', markersize=10, zorder=10)
    ax.plot(np.array(pts)[-1,0], np.array(pts)[-1,1], -1, 'o-', c='b', alpha=.5, markersize=7, zorder=10)
    
    # Plot Gradient Vector
    X, Y, Z = [pts[-1][0]], [pts[-1][1]], [-1]
    U, V, W = [-grad[0]], [-grad[1]], [0]
    ax.quiver(X, Y, Z, U, V, W, color='g')
    fig.savefig(TEMPFILE)
    plt.close()
    return im.fromarray(np.asarray(im.open(TEMPFILE)))

def loss(x):
    return np.sin(sum(x**3)) # change this

def gradient(x):
    return 2 * x * np.cos(sum(x**3)) # change this

def gradient_descent(x, y, init, learning_rate, epochs):
    images, losses, pts = [], [loss(init)], [init]
    for _ in range(epochs):
        grad = gradient(init)
        images.append(snap(x, y, pts, losses, grad))
        init = init - learning_rate * grad
        losses.append(loss(init))
        pts.append(init)
    return images

init = np.array([-1, -1]) # change this
learning_rate = 1.9587 # change this
x, y = np.meshgrid(np.arange(-2, 2, 0.1), np.arange(-2, 2, 0.5)) # change this
images = gradient_descent(x, y, init, learning_rate, 12)

images[0].save(
    'gradient_descent.gif',
    optimize=False,
    save_all=True,
    append_images=images[1:],
    loop=0,
    duration=500
)
```
