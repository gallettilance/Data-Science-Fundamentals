### Cleaning the data
Now that we have the data, we need to clean it. For example, some `cost_to_make` fields of some created cookies are missing. To resolve this, we can do many things: we can replace the missing data with the mean of the column, or we can get rid of the row entirely if the `cost_to_make` field is not set. 

```python
df = df.dropna()
df
```

To also make sure we removed null values, we can check the number of null values in each column using the `isnull()` function.

```python
df.isnull()
```

Next, lets check for duplicate rows using the `duplicated()` function. Then, remove those rows using `drop_duplicates()` function.

```python
df.duplicated()
df = df.drop_duplicates()
df
```

### Parsing the data
Now that we have the data, we could parse it to get the information we want. For example, we can check what types of cookies were made by using the `unique()` function on the `cookie` column.

```python
df["cookie"].unique()
```

We can also check the number of cookies made by using the `value_counts()` function on the `cookie` column.

```python
df["cookie"].value_counts()
```

Or maybe we don't like how long the names of the cookies are, so we can shorten them by using the `replace()` function on the `cookie` column.

For example, let's try changing `"bad berry bunny"` to `"bbb"`.

```python
df["cookie"] = df["cookie"].replace("bad berry bunny", "bbb")
df
```

We may even like the original names better, but we may want to get rid of the spaces. For example, we can change `"orange ozuna"` to `"orange_ozuna"`. Here, we will use the `str.replace()` function.

```python
df["cookie"] = df["cookie"].replace("orange ozuna", "orange_ozuna")
df
```

We may even just want to keep the first word of the cookie name. For example, we can change `"orange_ozuna"` to `"orange"`.

```python
for cookie in df["cookie"]:
    new_name = cookie.split()
    #print(new_name)
    df["cookie"] = df["cookie"].replace(str(cookie), str(new_name[0]))
df
```

Another thing that may come to mind is that maybe getting flour could cost more money due to inflation, so we have to adjust our `cost_to_make` values. We can do this by using the `apply()` function on the `cost_to_make` column.

```python
#Don't edit this method
def adjust_cost(cost):
    return cost + 0.5

df["cost_to_make"] = df.apply(lambda x: adjust_cost(x["cost_to_make"]), axis=1)
df

# df["cost_to_make"].apply(adjust)
```

### More complicated operations: Grouping, Filtering, Aggregating

Before trying out these complicated operations, lets first sort the sort the df by the radius of the cookies using the `sort_values()` function.

```python
# TODO: sort the df using sort_values(by='Column', ascending=False)
df = df.sort_values(by=["radius"], ascending=False)
df
```

We may also want to group data by certain attributes. This can be done by using `groupby()`. This method takes in a column name, and groups the data by the values in that column. For example, we can group the data by the `cookie` column.

```python
#TODO: group by cookie type
by_cookie = df.groupby("cookie")
by_cookie.first()
```

We can also group by multiple columns. For example, we can group the data by the `cookie` and `ingredients` columns.

```python
#TODO: group by cookie type and ingredients
by_2 = df.groupby(["cookie", "ingredients"])
by_2.head()
```

We may also want to filter the data. For example, we can filter the data to only show the cookies that have a radius greater than 4.3 cm. We can do this by indexing the DataFrame with a boolean expression.

```python
#TODO: filter using the boolean expression
df[df["radius"] > 4.3]
```

We may even want to use `groupby()` and filter idea together! For example, we can filter the data to only show the cookies that have a radius greater than 4.3 cm, and group the data by the `cookie` column.

```python
#TODO: filter the data using the boolean expression
# then group by cookie column
two_again = df[df["radius"] > 4.3].groupby("cookie")
two_again.head()
```

We may also want to aggregate the data. For example, we can aggregate the data by looking at the ratio of calories to radius. We can do this by using indexing and the `apply()` function.

```python
#TODO: add a column to the dataframe that is the
# calories per radius
def f(column1, column2):
    return column1 / column2

df["cal_to_rad"] = df.apply(lambda x: f(x["calories"], x["radius"]), axis=1)
df
```

Or we can just get rid of this column if we find it useless. We can do this by using the `drop()` function or indexing.

```python
#TODO: drop the created column
df = df.drop(columns=["cal_to_rad"])
df
```
